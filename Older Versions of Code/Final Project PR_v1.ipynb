```python
def get_file(csv_files):
    
    """Given a filename which has a comma separated (.csv) filetype,return a N dimensional 
    array (numpy.ndarray) containing the unique AircraftHex values.
    
    We load the file as a pandas dataframe having no headers and load only 1 relevant column 
    needed for analysis called (AircraftHex) which corresponds to column number 4 in the csv 
    file. We then rename the column 4 to AIrcraftHex and create an ndarray having unique 
    AircraftHex values.
    
    :param airplane_datafile: The csv file to load
    :return: a N dimensional array (numpy.ndarray) containing the unique AircraftHex values
    """
    data_file = pd.read_csv(csv_files,usecols=['CMPLNT_NUM','CMPLNT_FR_DT','BORO_NM']) # Import only necessary columns from the dataset
    #arrest_df = arrest_df.rename(columns={4: 'Crime Description'})
    return data_file
#Step 1 complete
```

```python
NYC_Complaint_Data = get_file(Complaints) 
NYC_Complaint_Data #106 seconds to run
```

```python
NYC_Complaint_Data['Complaint_Filed_Year'] = NYC_Complaint_Data.CMPLNT_FR_DT.str[-4:]
NYC_Complaint_Data
```

```python
NYC_Complaint_Data = NYC_Complaint_Data.dropna(subset=['Complaint_Filed_Year'])
NYC_Complaint_Data = NYC_Complaint_Data.astype({'Complaint_Filed_Year': 'int64'})
NYC_Complaint_Data = NYC_Complaint_Data[NYC_Complaint_Data['Complaint_Filed_Year'] > 2005]
NYC_Complaint_Data
```

```python
Crime_count = NYC_Complaint_Data.groupby(['BORO_NM','Complaint_Filed_Year']).agg({'CMPLNT_NUM': ['count']}) #https://pandas.pydata.org/pandas-docs/version/0.23.1/generated/pandas.core.groupby.DataFrameGroupBy.agg.html
Crime_count
```

```python
#replacing Borough initials with complete names
NYC_Arrest_Data = NYC_Arrest_Data.replace({'ARREST_BORO': 'K'}, 'BROOKLYN')
NYC_Arrest_Data = NYC_Arrest_Data.replace({'ARREST_BORO': 'M'}, 'MANHATTAN')
NYC_Arrest_Data = NYC_Arrest_Data.replace({'ARREST_BORO': 'S'}, 'STATEN ISLAND')
NYC_Arrest_Data = NYC_Arrest_Data.replace({'ARREST_BORO': 'B'}, 'BRONX')
NYC_Arrest_Data = NYC_Arrest_Data.replace({'ARREST_BORO': 'Q'}, 'QUEENS')
NYC_Arrest_Data
```

```python
NYC_Arrest_Data['Arrest_Year'] = NYC_Arrest_Data.ARREST_DATE.str[-4:]
NYC_Arrest_Data
```

```python
NYC_Arrest_Data = NYC_Arrest_Data.dropna(subset=['Arrest_Year'])
NYC_Arrest_Data = NYC_Arrest_Data.astype({'Arrest_Year': 'int64'})
NYC_Arrest_Data = NYC_Arrest_Data[NYC_Arrest_Data['Arrest_Year'] > 2005]
NYC_Arrest_Data
```

```python
Arrest_count = NYC_Arrest_Data.groupby(['ARREST_BORO','Arrest_Year']).agg({'ARREST_KEY': ['count']}) #https://pandas.pydata.org/pandas-docs/version/0.23.1/generated/pandas.core.groupby.DataFrameGroupBy.agg.html
Arrest_count
```

```python
result = pd.concat([Crime_count, Arrest_count], axis=1, join='inner')
```
